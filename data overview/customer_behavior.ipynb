```python
import pandas as pd
```

```python
df = pd.read_csv("behavior_customer.csv")
```

```python
df.head(10)
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df['Review Rating'].fillna(df['Review Rating'].median())
```

```python
df.isnull().sum()
```

```python

df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
```

```python
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[0,18,30,45,60,100],
    labels=['Teen','Young','Adult','Middle','Senior'])
# # create a new column age_group
# labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
# df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['purchase_level'] = pd.cut(
    df['purchase_amount'],
    bins=[0,50,100,150,1000],
    labels=['Low','Medium','High','Premium']
)
```

```python
df[['purchase_amount','purchase_level']].head(10)
```

```python

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python

df = df.drop('promo_code_used', axis=1)
```
